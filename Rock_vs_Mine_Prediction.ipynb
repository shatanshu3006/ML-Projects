Importing the Dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

Data Collection And Data Processing


```python
# Loading the Dataset into the pandas DataFrame

#Pandas DataFrame is a widely used data structure which works with a 
#two-dimensional array with labeled axes (rows and columns). 
#DataFrame is defined as a standard way to store data that has two different indexes,
#i.e., row index and column index.

sonar_data=pd.read_csv('/content/Copy of sonar data.csv',header=None) #naming the file as sonar_data
```

```python
sonar_data.head(5)        #this method shows the first five rows for the choosen file
```

```python
# number of rows and columns
sonar_data.shape
```

```python
sonar_data.describe() # describes the statistical measures of the data
```

```python
sonar_data[60].value_counts() # gives how many rock and mine data counts are there
# 60 is for the 60th column (0 indexed) that gives the number of rocks and mines
```

M --> Mine

R --> Rocks

```python
sonar_data.groupby(60).mean()
```

```python
#separating the data and the labels
x= sonar_data.drop(columns=60,axis=1)
y=sonar_data[60]
```

```python
print(x)
print(y)
```

Training And Test Data

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(x.shape,x_test.shape,x_train.shape)
```

Model Training --> Logistic Regression Model

```python
print(x_train)
print(y_train)
```

```python
#already imported the logiistic regression 
model= LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(x_train,y_train)
```

Model Evaluation


```python
#accuracy on training data
#accuracy of training data is generally higher than that of the test data as machine has already seen that
#since the size of this data is small, the accuracy will be low
#accuracy of around 70-80% is generally considered good

x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train) # the prediction on training data of x is checked for accuracy against y_train that is the real dataset for the same.
```

```python
print('Accuracy on training data : ',training_data_accuracy)
```

```python
#accuracy on test data

x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on test data : ',test_data_accuracy)
```
